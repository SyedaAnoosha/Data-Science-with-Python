## Missing Values

- Drop missing values if they are 5% or less than of total values
- If not then impute mean, median or mode

```python
unemployment.isna().sum()
```

```python
# Printing only those rows where data is missing
unemployment.loc[unemployment['continent'].isna(),:]
```

```python
limit = len(unemployment) * 0.05 # 5% check
limit
```

```python
cols_to_drop = unemployment.columns[unemployment.isna().sum() <= limit]
cols_to_drop
```

```python
unemployment.dropna(subset = cols_to_drop, inplace = True)
```
