```python
import pandas as pd
import matplotlib.pyplot as plt

wine_reviews = pd.read_csv("Dataset/winemag-data-130k-v2.csv", index_col=0)
wine_reviews
```

```python
wine_grouped = wine_reviews.groupby("country")["price"].max()
wine_grouped
```

```python
wine_grouped.plot(kind='bar', title="Maximum Priced Wine in Particular Country")
plt.show()
```

```python
country_reviews = wine_reviews.groupby("country")["country"].value_counts()
country_reviews.plot(kind='bar')
```
