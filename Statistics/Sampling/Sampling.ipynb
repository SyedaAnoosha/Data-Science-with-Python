# Sampling

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
spotify_population = pd.read_feather('Datasets/spotify_2000_2020.feather')
spotify_population.head()
```

```python
spotify_population["artists"] = spotify_population['artists'].str.replace("['","")
spotify_population["artists"] = spotify_population['artists'].str.replace("']","")
spotify_population.head()
```

```python
spotify_population[spotify_population['artists']=='Taylor Swift'].sort_values(by="year", ascending=False).head()
```

```python
# Random sample data of 1000 rows 
spotify_sample = spotify_population.sample(n=1000)
spotify_sample.head()
```

```python
mean_dur_pop = spotify_population["duration_minutes"].mean()
mean_dur_samp = spotify_sample["duration_minutes"].mean()

print(mean_dur_pop)
print(mean_dur_samp)
```

```python
loudness_pop = pd.Series(spotify_population["loudness"])
loudness_samp = loudness_pop.sample(n=100)

print(np.mean(loudness_pop))
print(np.mean(loudness_samp))
```

## Convenience Sampling
The data collected by the easiest method is called convenience sampling and is often prone to sample bias. 

Imagine you need to collect data from people all over the city. You decided to go to each house and collect the data in person, but that would be time-consuming. Instead, you decide to send out a survey in the mail. This is a convenient way to collect data, but it can lead to sample bias. This is because people who do not have a mailing address, such as homeless people are less likely to receive the survey. As a result, your data will not be representative of the entire population of the city.
