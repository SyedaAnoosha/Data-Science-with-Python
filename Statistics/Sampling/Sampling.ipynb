## Convenience Sampling
The data collected by the easiest method is called convenience sampling and is often prone to sample bias. 

Imagine you need to collect data from people all over the city. You decided to go to each house and collect the data in person, but that would be time-consuming. Instead, you decide to send out a survey in the mail. This is a convenient way to collect data, but it can lead to sample bias. This is because people who do not have a mailing address, such as homeless people are less likely to receive the survey. As a result, your data will not be representative of the entire population of the city.

## Simple Random Sampling

Every member of the population has an equal chance of being selected.

```python
attrition = pd.read_feather('Datasets/attrition.feather')
attrition.head()
```

```python
attrition_samp = attrition.sample(n=70, random_state=18900217)
attrition_samp
```

## Systematic Sampling

Every member of population is listed with a number. Instead of randomly generating numbers, individuals are chosen at regular intervals.

```python
sample_size = 70
pop_size = len(attrition)

interval = pop_size // sample_size

attrition_sys_samp = attrition.iloc[::interval]
attrition_sys_samp.head()
```

## Stratified Random Sampling

It involves dividing the population into subpopulations that may differ in important ways. This will make it easy to generalize the results about sampling data.

For example following I have subsetted the whole population on the basis of country_of_origin and then sampling from that dataset.

```python
coffee_ratings = pd.read_feather('Datasets/coffee_ratings_full.feather')
coffee_ratings.head()
```

```python
# Sampling on the basis of most popular country_of_origin

top_counted_countries = ["Mexico","Colombia","Guatemala","Brazil","Taiwan","United States (Hawaii)"]
```

```python
coffee_ratings["country_of_origin"].value_counts(normalize=True).head(6)
```

```python
# Subsetting data on the basis of country_of_origin
coffee_ratings_top = coffee_ratings[coffee_ratings["country_of_origin"].isin(top_counted_countries)]

# Sampling data from top 6 countries sample
coffee_ratings_top_samp = coffee_ratings_top.sample(frac = 0.1, random_state = 12345)
coffee_ratings_top_samp.head()
```

```python
coffee_ratings_top_samp["country_of_origin"].value_counts(normalize=True) 
```

####  Proportional stratified sampling 

```python
# Proportion of employees by Education level
education_counts_pop = attrition['Education'].value_counts(normalize=True)
education_counts_pop
```

```python
# Proportional stratified sampling for 40% of each Education group
attrition_strat = attrition.groupby('Education')\
	.sample(frac=0.4, random_state=2022)

attrition_strat.head()
```

```python
# Calculate the Education level proportions from attrition_strat
education_counts_strat = attrition_strat['Education'].value_counts(normalize=True)

education_counts_strat
```

#### Equal counts stratified sampling

```python
# Get 30 employees from each Education group
attrition_eq = attrition.groupby('Education')\
	.sample(n=30, random_state=2022)
attrition_eq.head()
```

```python
education_counts_eq = attrition_eq["Education"].value_counts(normalize=True)
education_counts_eq
```
