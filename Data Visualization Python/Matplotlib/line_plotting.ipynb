```python
import matplotlib.pyplot as plt

year = [1950, 1970, 1990, 2010, 2020]
pop = [2.3, 3.5, 5.6, 6.9, 8.9]
```

### Matplotlib plot()
Conect points

```python
plt.plot(year, pop)
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```

### Matplotlib scatter()
Does not connect points

```python
plt.scatter(year, pop)
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```
